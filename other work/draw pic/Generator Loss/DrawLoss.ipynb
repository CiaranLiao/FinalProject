```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have the CSV files named 'a.csv', 'b.csv', 'c.csv', and 'd.csv'
# and each CSV file has columns 'Step' for the x-axis and 'Value' for the y-axis.

# Read the CSV files
df_a = pd.read_csv('starfish.csv')
df_b = pd.read_csv('starfish_edge_gray.csv')
df_c = pd.read_csv('starfish_edge_resume.csv')
df_d = pd.read_csv('starfish_edge_RGB.csv')

# Plotting all the losses in one graph
plt.figure(figsize=(5, 5))

plt.plot(df_a['Step'].values, df_a['Value'].values, label='Loss starfish')
plt.plot(df_b['Step'].values, df_b['Value'].values, label='Loss edge_gray')
plt.plot(df_d['Step'].values, df_d['Value'].values, label='Loss edge_RGB')
plt.plot(df_c['Step'].values, df_c['Value'].values, label='Loss edge_resume')

plt.title('Generator Loss')
plt.xlabel('Step')
plt.ylabel('Loss Value')
plt.legend(loc='center left',bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.show()
```
