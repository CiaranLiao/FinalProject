```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取 CSV 文件
df_dloss = pd.read_csv('Dloss.csv')
df_gloss = pd.read_csv('Gloss.csv')
df_fs = pd.read_csv('fs.csv')
df_rs = pd.read_csv('rs.csv')
df_r1 = pd.read_csv('r1.csv')

# 创建三个子图
fig, axs = plt.subplots(3, 1, figsize=(10,8))  # 3个图表，每个图表的宽度为10，总高度为15

# Dloss 和 Gloss 在同一张图
axs[0].plot(df_dloss['Step'].values, df_dloss['Value'].values, label='D Loss')
axs[0].plot(df_gloss['Step'].values, df_gloss['Value'].values, label='G Loss', color='orange')
axs[0].set_title('Discriminator and Generator Loss')
axs[0].set_xlabel('Step')
axs[0].set_ylabel('Loss')
axs[0].legend()
axs[0].grid(True)

# Fake Score 和 Real Score 在同一张图
axs[1].plot(df_fs['Step'].values, df_fs['Value'].values, label='Fake Score', color='green')
axs[1].plot(df_rs['Step'].values, df_rs['Value'].values, label='Real Score', color='red')
axs[1].set_title('Fake and Real Score')
axs[1].set_xlabel('Step')
axs[1].set_ylabel('Score')
axs[1].legend()
axs[1].grid(True)

# R1 Penalty 单独一张图
axs[2].plot(df_r1['Step'].values, df_r1['Value'].values, label='R1 Penalty', color='purple')
axs[2].set_title('R1 Penalty')
axs[2].set_xlabel('Step')
axs[2].set_ylabel('Penalty')
axs[2].legend()
axs[2].grid(True)

# 调整子图间的间距
plt.tight_layout()
plt.show()
```
